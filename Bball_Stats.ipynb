## Cleaning up the data

Check for NaN values

```python
df_3pt_era.isna().sum()
```

After checking for NaN values, I notice the columns containing NaN values are those with percentages, where likely divided zero by zero. This is because there are samples where players didn't even attempt shots at all, and thus recorded no data. 

USG% (Usage %) and PER (Player Efficiency Rating) are calculations that could involve dividing by 0 as well, so we'll treat them similarly and replace all above columns containing NaN values with 0.0.

```python
cols_with_nan = ['PER', 'TS%', 'USG%', 'FG%', '3P%', '2P%', 'eFG%', 'FT%']
df_3pt_era[cols_with_nan] = df_3pt_era[cols_with_nan].fillna(0)
df_3pt_era
```

In previous readouts of the data, I saw index 5731, Dennis Awtrey, recorded 0 3PA and 0 3PM, which would yield a NaN value in 3P%.

Testing now to verify the code worked to replace

```python
df_3pt_era.loc[5731]
```

```python
df_3pt_era.head()
```

```python
df_3pt_era.drop(columns='Unnamed: 0', inplace=True)
df_3pt_era
```

```python
df_3pt_era.columns
```

Looking at the above columns, I can create a list of those I want to format. I will select all counting stat categories that are strictly whole numbers and thus would never be fractional and also include categories like "Year" and "Age".

```python
cols_to_format = ['Year', 'Age', 'G', 'MP', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT', 'FTA', 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']

df_3pt_era[cols_to_format] = df_3pt_era[cols_to_format].astype(int).apply(lambda x: x.replace('.0', ''))
df_3pt_era
```

Since I'm only interested in his total stats for the season I could filter or delete the rows, but what about the other players who played for multiple teams? That would be cumbersome. Let's use the *groupby* and *agg* functions to only take the maximum values for an observation.

We want to keep the player name as a column, though, so we can use it in our plot function as a parameter. So, we'll create a duplicate column to group by first.

```python
# duplicating player's name and grouping by it so name serves as index while
# preserving column to pass into our plot function later

df_3pt_era['Player_Name'] = df_3pt_era['Player']
by_player = df_3pt_era[df_3pt_era.Year == 2017].groupby('Player_Name').agg('max')


# viewing dataframe filtering the year 2017 

by_player[by_player.Year == 2017]
```

Now we can group by the 'Player Name' column and still keep the 'Player' column for our plot function.

Let's check our work.

```python
by_player.duplicated().sum()
```

```python
fig = px.scatter(by_player.query('Year == 2017'), x='USG%', y='PTS',
	         size='FGA', color='Pos',
                 hover_name='Player', log_x=False, size_max=60)
fig.show()
```

That graph looks pretty busy. I want to filter some of the data to limit it to more prominent players. Let's look to explore the data and create some sort of threshold involving a combination of things.

```python
print(by_player[by_player.Year == 2017].PTS.describe())
print(by_player[by_player.Year == 2017].G.describe())
print(by_player[by_player.Year == 2017].FGA.describe())
print(by_player[by_player.Year == 2017].FTA.describe())
print(by_player[by_player.Year == 2017].VORP.describe())
print(by_player[by_player.Year == 2017].BPM.describe())
print(by_player[by_player.Year == 2017].MP.describe())
```

Given my comfort level with fantasy basketball (read: pays a lot of attention to stats) and after reviewing above descriptive statistics from key statistical categories, I've selected thresholds for several catergories select for the more prominent players.

```python
# note conditional chain using points, minutes played, and box-score plus-minus 
# along witht the year filter of 2017

fig = px.scatter(by_player.query('Year == 2017 & PTS > 750 & MP > 1984 & BPM > .75'), x='USG%', y='PTS',
	         size='BPM', color='Pos',
                 hover_name='Player', hover_data=['FGA', 'FG%', 'VORP'], log_x=False, size_max=60)

fig.update_layout(
    title='Top Players for 2017'
)

fig.show()
```

The first thing you notice with this less cluttered graph is Russell Westbrook, leading the league in usage and winning the scoring title in 2017 (the man averaged a triple double, first time in 55 years since Oscar Robinson). It's no surprise he won the MVP this year. Nice job, Brodie! 

However, the datapoints are all **totals** for the 2017 season. Some guys missed games due to injury, others played fewer games because their teams did not make the playoffs, etc. 

While Westbrook's dependability contributed to his raw volume and MVP justification, let's investigate from a different vantage point and explore **per-game** averages.

```python
by_player[by_player.Year == 2017][by_player.PTS > 750][by_player.MP > 1984][by_player.BPM > .75].G.describe()
```

Above you can see some variance in the total games played. With an average games played of 75 in an 82-game regular season (plus additional playoff games), this is worth our time to keep investigating.

If I want the **per game** value for the columns of interest, I'll have to create new **per game columns** by dividing all of the counting stat columns by games played. Let's see how this changes our data with the same filters applied.

```python
by_player[by_player.Year == 2017].columns
```

```python
# run a quick for loop to format

counting_stat_cols = ['MP', 'FG', 'FGA', '3P', '3PA', '2P', '2PA', 'FT', 'FTA',
                 'ORB', 'DRB', 'TRB', 'AST', 'STL', 'BLK', 'TOV', 'PF', 'PTS']
per_game_cols = []
for col in counting_stat_cols:
  col = f'{col}_PG'
  per_game_cols.append(col)


# another for loop to tie in the new per game columns by dividing counting stat
# totals by each player's total games played

for col in range(0, len(counting_stat_cols)):
  by_player[per_game_cols[col]] = by_player[counting_stat_cols[col]] / by_player.G


# check resulting dataframe

by_player
```

```python
print(by_player.columns)
```

```python
# formatting values in newly calculated per game columns

by_player[per_game_cols] = by_player[per_game_cols].round(3)
by_player[per_game_cols]
```

The new columns have been added and formatted. Let's see how it looks versus the season totals bubble chart using the same criteria.
